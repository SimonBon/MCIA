```python
from pathlib import Path
from tifffile import imread
from tqdm import tqdm
import pandas as pd
import numpy as np

base = Path('/home/simon_g/isilon_images_mnt/10_MetaSystems/MetaSystemsData/_simon/data/cHL_data/zenodo_files/cHL_CODEX')
```

```python
samples = list((base / 'raw_image').glob('*'))
print(samples)
```

```python
sample_dict = dict()
for i, sample in enumerate(samples):

    sample_id = sample.stem
    files = list(sample.glob('*'))
    print(files)
        
    files = list(sample.glob('*'))
    marker_image_dict = dict()
    pbar = tqdm(files, desc=f"{i}/{len(samples)}")
    for f in pbar:
        marker = str(f.stem).replace(' ', '')
        if marker == 'Segmentation':
            print('Skipping Segmentation')
            continue
        #print(marker)
        image = imread(f).T
        image = image / 65535.
        print(image.shape)
        marker_image_dict[marker] = image
        
    segmentation = imread(base / f'segmentation/{sample_id}/cHL_CODEX_segmentation.tiff').T
    marker_image_dict['segmentation'] = segmentation
        
    sample_dict[sample_id] = marker_image_dict
```

```python
annotations_df = pd.read_csv(base / 'annotation_csv/cHL_CODEX_annotation.csv')
```

```python
annotations_df.columns
```

```python
cell_labels = annotations_df.cellType.to_numpy()
cell_labels
```

```python
X = annotations_df.X_cent.to_numpy()
Y = annotations_df.Y_cent.to_numpy()
ID = np.ones_like(X)
```

```python
identifier = np.stack((X, Y, ID), axis=1)
identifier.shape
```

```python
identifier
```

```python
sample_ids = [int(s) for s in sample_dict.keys()]
mask = np.isin(identifier[:, 2], sample_ids)
_id = identifier[mask]
_cell_labels = cell_labels[mask]
_id.shape, _cell_labels.shape
```

```python
sample_dict['1']['PD-1'].shape, _id[:, 0].max(), _id[:, 1].max()
```

```python
from collections import Counter
import numpy as np

# Example:
mask_values = sample_dict['1']['segmentation'][_id[:, 0], _id[:, 1]]

# Count occurrences of each mask value
counts = Counter(mask_values)

# Find mask values that occur exactly once and are not zero
valid_mask_values = {val for val, cnt in counts.items() if cnt == 1 and val != 0}

# Create a boolean mask for entries to keep
keep_mask = np.array([val in valid_mask_values for val in mask_values])

# Apply the mask to filter all arrays
mask_values_filtered = mask_values[keep_mask]
_id_filtered = _id[keep_mask]
_cell_labels_filtered = _cell_labels[keep_mask]

print(f"Original length: {len(mask_values)}, filtered length: {len(mask_values_filtered)}")
len(_id_filtered), len(_cell_labels_filtered), len(mask_values_filtered)
```

```python
from pathlib import Path
imaging_data_folder = Path('/home/simon_g/isilon_images_mnt/10_MetaSystems/MetaSystemsData/_simon/data/MCI_data')
dataset_name = 'CODEX_cHL'

dataset_folder = imaging_data_folder / dataset_name
dataset_folder.mkdir(exist_ok=True)

target_images_folder = dataset_folder / 'images'
target_images_folder.mkdir(exist_ok=True)

target_masks_folder = dataset_folder / 'masks'
target_masks_folder.mkdir(exist_ok=True)
```

```python
import numpy as np
import tifffile

marker_list = None
for sample, images_dict in sample_dict.items():
    
    
    if marker_list is None:
        marker_list = [m for m in images_dict.keys() if m != "segmentation"]
        marker_list = list(marker_list)

    images = np.array([images_dict[m] for m in marker_list]).transpose(1,2,0)
    masks = np.array(images_dict['segmentation'])
    
    tifffile.imwrite(target_images_folder / f'{sample}_image.tif', images)
    tifffile.imwrite(target_masks_folder / f'{sample}_masks.tif', masks)
    
annotations = dict(
    DIM1=_id_filtered[:, 0],
    DIM2=_id_filtered[:, 1],
    SampleID=_id_filtered[:,2],
    annotation=_cell_labels_filtered,
    mask_value=mask_values_filtered
) 

annotations = pd.DataFrame(annotations)

annotations.to_csv(dataset_folder / 'annotations.csv', index=False)

panel = pd.DataFrame(marker_list, columns=['Label'])
panel.to_csv(dataset_folder / 'marker_names.csv')
    
```

```python
import h5py
import numpy as np

out_path = "/home/simon_g/isilon_images_mnt/10_MetaSystems/MetaSystemsData/_simon/src/MCA/dataset/cHL_CODEX.h5"

with h5py.File(out_path, "w") as f:
    marker_list = None

    for sample, images_dict in sample_dict.items():
        if marker_list is None:
            marker_list = [m for m in images_dict.keys() if m != "segmentation"]
            marker_list = list(marker_list)

            str_dtype = h5py.string_dtype(encoding="utf-8")
            f.create_dataset(
                name="MARKERS",
                data=np.array(marker_list, dtype=str_dtype)
            )
            
        images = np.array([images_dict[m] for m in marker_list]).transpose(1,2,0)
        masks = np.array(images_dict['segmentation'])
        print(images.shape, masks.shape)

        sample = f.create_group(name=str(sample))
        sample.create_dataset(name='IMAGES', data=images, chunks=(64, 64, 1), compression=3)
        sample.create_dataset(name='MASKS', data=masks, chunks=(64, 64))
                
    f.create_dataset(name='INDEXER', data=_id)
    f.create_dataset(name='IMAGE_KEYS', data=np.array(list(sample_dict.keys()), dtype=str_dtype))
    f.create_dataset(name='ANNOTATIONS', data=_cell_labels, dtype=str_dtype)
```
